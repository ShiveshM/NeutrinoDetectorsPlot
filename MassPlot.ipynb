```python
%matplotlib inline

import numpy as np
import matplotlib.pyplot as plt
from ipywidgets import interact, interactive, fixed, interact_manual
import ipywidgets as widgets
import adjustText

def PlotStuff(pl1=True,pl2=True,pl3=True,pl4=True,pl5=True, pl6=True):
    #
    # water cherenkov detectors
    #

    pl1 = bool(pl1)
    pl2 = bool(pl2)
    pl3 = bool(pl3)
    pl4 = bool(pl4)
    pl5 = bool(pl5)
    pl6 = bool(pl6)
    
    rotationvalue=40
    
    # amanda
    amandaDetectors = ["AMANDA-A", "AMANDA-B10", "AMANDA-II"]
    amandaYears = [1996, 1997, 2000]
    amandaSpatRes = [2.62, 4.93, 6.03]
    amandaMasses = [1.14e08, 5.89e08, 3.78e10]

    # baikal
    baikalDetectors=["Baikal NT-200", "Baikal NT-200+"]#, "Baikal-GVD"]
    baikalYears = [1993, 2005]#, 2020]
    baikalSpatRes = [2.12, 2.12]#, 6.98]
    baikalMasses = [2e08, 2e08]#, 907184740000]

    # icecube
    icecubeDetectors = [ "IceCube", "IC-DeepCore"]#, "IC-Pingu"]
    icecubeYears = [2010,2011]#, 2020]
    icecubeSpatRes = [13.99,3.92]#, 1.91]
    icecubeMasses = [907184740000, 27215542200]#, 5443108440]

    # super k
    superkDetectors = ["SK-I", "SK-II", "SK-III", "SK-IV"]
    superkYears = [1996, 2002, 2006, 2008]
    superkSpatRes = [ 0.16, 0.23, 0.16, 0.16]
    superkMasses = [45359237,45359237,45359237,45359237]

    # others
    wCherenkovDetectors = ["ANNIE-I", "ANTARES", "Hyper-K", "Kamiokande", "MiniBooNE"]
    wCherenkovYears = [2016, 2008, 2025, 1983, 2002]
    wCherenkovSpatRes = [0.182, 4.65, 0.22, 0.21,0.09]
    wCherenkovMasses=[23586, 3.5e10, 335658354, 2721544,7.42e05]
    wCherenkov2Detectors = ["SNO"]
    wCherenkov2Years = [1999]
    wCherenkov2SpatRes = [0.15]
    wCherenkov2Masses = [30000]

    # Tracking detectors
    trackingDetectors = ["AGS Spark Chamber","KARMEN", "MINERvA", "MINOS-FD", "NOvA-FD", "OPERA", "DONuT", "NuTeV"]
    trackingYears = [1961,1990, 2010, 2005, 2014, 2008, 1997, 1996]
    trackingSpatRes = [0.03,0.06, 0.003, 0.0256, 0.0450, 0.00005,0.0019, 0.0003]
    trackingMasses = [9.1e03, 62660, 5.5e03, 4.9e06, 1.27e07, 1.63e6, 260, 7e05]
    tracking2Detectors = ["SciBooNE","Goesgen","Palo Verde"]
    tracking2Years = [2003, 1981,1998]
    tracking2SpatRes = [0.0201, 0.1,0.03]
    tracking2Masses = [1.36e04, 3.24e2,1e4]

    #TPCs
    tpcDetectors = ["ICARUS","MicroBooNE", "SBND", "DUNE"]
    tpcYears = [2010,2015, 2020, 2025]
    tpcSpatRes = [0.00131,0.00153, 0.00192, 0.00206]
    tpcMasses = [431820,80739, 101605, 36287390]
    tpc2Detectors = ["ArgoNEUT", "NOMAD-TPC", "ND280-TPC"]
    tpc2Years = [2009, 1995, 2009]
    tpc2SpatRes = [0.00157, 0.0003, 0.0007]
    tpc2Masses = [690,2.5e6, 2.5e3 ]
    
    # Bubble Chambers
    bcDetectors =  ["ANL", "FERMILAB-15FT", "Gargamelle"]
    bcYears = [1979, 1973.5, 1976]
    bcSpatRes = [2e-6, 2e-6, 2e-6]
    bcMasses = [4.6, 4.6, 18560]
    bc2Detectors = ["BNL", "BEBC"]
    bc2Years = [1974, 1972.5]
    bc2SpatRes = [2e-6, 2e-6]
    bc2Masses = [4.6, 4.6]
    
    # Unsegmented liquid scintilator
    scinDetectors = ["KamLAND","Borexino", "CHOOZ", "Hanford", "Daya Bay", "JUNO" , "SNO+"]
    scinYears = [2002,2007, 1997, 1953, 2011, 2020, 2017]
    scinSpatRes = [1.8e-2,1.39e-2, 1.755e-2, 2.6e-2, 1.755e-2, 1.69e-2, 1.31e-2]
    scinMasses = [1e06,2.8e05, 4.54e03, 2.4e02, 1.45e05, 20000000 ,780000]
    scin2Detectors = [ "RENO", "LSND", "Double CHOOZ"]
    scin2Years = [ 2011, 1993, 2015]
    scin2SpatRes = [ 1.59e-2,0.014,0.019]
    scin2Masses = [ 3e4,1.52e05,3e05]
    
    # radiochemical
    radDetectors = ["GALLEX/GNO", "Homestake"]
    radYears = [1991, 1970]
    radSpatRes = [3.77, 7.24]
    radMasses = [27215, 605665]
    rad2Detectors = ["SAGE"]
    rad2Years = [1989]
    rad2SpatRes = [2.02]
    rad2Masses = [45359]
    
    fig = plt.figure(figsize=(20*4/3,20))
    ax = fig.add_subplot(111)
    if pl1 is True: 
        ax.plot(wCherenkovYears, wCherenkovMasses, color='mediumseagreen', label='Cherenkov', linestyle="", marker='o',markersize=20)
        ax.plot(wCherenkov2Years, wCherenkov2Masses, color='mediumseagreen', linestyle="", marker='o', markersize=20)
        ax.plot(amandaYears, amandaMasses, color='mediumseagreen', marker='o', ls='--',markersize=20)
        ax.plot(baikalYears, baikalMasses, color='mediumseagreen', marker='o', ls='--',markersize=20)
        ax.plot(icecubeYears, icecubeMasses, color='mediumseagreen', marker='o', ls='--',markersize=20)
        ax.plot(superkYears, superkMasses, color='mediumseagreen', marker='o', ls='--',markersize=20)
            
        ax.annotate("AMANDA", (amandaYears[2], amandaMasses[2]), color='seagreen', fontsize=24, rotation=rotationvalue, va='bottom', ha='left')
        #ax.annotate("BAIKAL-GVD", (baikalYears[2], baikalMasses[2]), color='seagreen', fontsize=24, rotation=rotationvalue, va='bottom', ha='left')
        ax.annotate("IceCube", (icecubeYears[0], icecubeMasses[0]), color='seagreen', fontsize=24, rotation=rotationvalue, va='bottom', ha='left')
        ax.annotate("DeepCore", (icecubeYears[1], icecubeMasses[1]), color='seagreen', fontsize=24, rotation=rotationvalue, va='bottom', ha='left')
        #ax.annotate("PINGU", (icecubeYears[2], icecubeMasses[2]), color='seagreen', fontsize=24, rotation=rotationvalue, va='bottom', ha='left')
        ax.annotate("SK", (superkYears[0], superkMasses[0]), color='seagreen', fontsize=24, rotation=rotationvalue, va='bottom', ha='left')
        
        for i, txt in enumerate(wCherenkovDetectors):
            ax.annotate(txt, (wCherenkovYears[i], wCherenkovMasses[i]), color='seagreen', fontsize=24, rotation=rotationvalue, va='bottom', ha='left')
            
        for i, txt in enumerate(wCherenkov2Detectors):
            ax.annotate(txt, (wCherenkov2Years[i], wCherenkov2Masses[i]), color='seagreen', fontsize=24, rotation=rotationvalue, va='top', ha='right')
    
    if pl2 is True:
        ax.plot(scinYears, scinMasses, color='coral', label='Unseg. Scint.', linestyle="", marker='o',markersize=20)
        ax.plot(scin2Years, scin2Masses, color='coral', linestyle='', marker='o',markersize=20)
        
        for i, txt, in enumerate(scinDetectors):
            ax.annotate(txt, (scinYears[i], scinMasses[i]), color='orange', fontsize=24, rotation=rotationvalue, va='bottom', ha='left')
    
        for i, txt, in enumerate(scin2Detectors):
            ax.annotate(txt, (scin2Years[i], scin2Masses[i]), color='orange', fontsize=24, rotation=rotationvalue, va='top', ha='right')
    
    if pl3 is True: 
        ax.plot(trackingYears, trackingMasses, color='dimgray', label='Seg. Tracker',linestyle="", marker='o',markersize=20)
        ax.plot(tracking2Years, tracking2Masses, color='dimgray', linestyle='', marker='o',markersize=20)
        
        for i, txt, in enumerate(trackingDetectors):
            ax. annotate(txt, (trackingYears[i], trackingMasses[i]), color='dimgray', fontsize=24, rotation=rotationvalue, va='bottom', ha='left')
    
        for i, txt, in enumerate(tracking2Detectors):
            ax. annotate(txt, (tracking2Years[i], tracking2Masses[i]), color='dimgray', fontsize=24, rotation=rotationvalue, va='top', ha='right')
    
    if pl4 is True: 
        ax.plot(tpcYears, tpcMasses, color='dodgerblue', label='TPC',linestyle="", marker='o',markersize=20)
        ax.plot(tpc2Years, tpc2Masses, color='dodgerblue', linestyle="", marker='o',markersize=20)
        
        for i, txt, in enumerate(tpcDetectors):
            ax. annotate(txt, (tpcYears[i], tpcMasses[i]), color='deepskyblue', fontsize=24, rotation=rotationvalue, va='bottom', ha='left')
        
        for i, txt, in enumerate(tpc2Detectors):
            ax. annotate(txt, (tpc2Years[i], tpc2Masses[i]), color='deepskyblue', fontsize=24, rotation=rotationvalue, va='top', ha='right')    
        
    if pl5 is True:
        ax.plot(bcYears, bcMasses, color='palevioletred', label='Bubble Chamber', linestyle="", marker='o',markersize=20)
        ax.plot(bc2Years, bc2Masses, color='palevioletred', linestyle="", marker='o',markersize=20)
        
        for i, txt, in enumerate(bcDetectors):
            ax. annotate(txt, (bcYears[i], bcMasses[i]), color='hotpink', fontsize=24, rotation=rotationvalue, va='bottom', ha='left')
   
        for i, txt, in enumerate(bc2Detectors):
            ax. annotate(txt, (bc2Years[i], bc2Masses[i]), color='hotpink', fontsize=24, rotation=rotationvalue, va='top', ha='right')
            
    if pl6 is True:
        ax.plot(radYears, radMasses, color='blueviolet', label='Radiochemical', linestyle="", marker='o',markersize=20)
        ax.plot(rad2Years, rad2Masses, color='blueviolet', linestyle="", marker='o',markersize=20)
        
        for i, txt, in enumerate(radDetectors):
            ax. annotate(txt, (radYears[i], radMasses[i]), color='mediumorchid', fontsize=24, rotation=rotationvalue, va='bottom', ha='left')

        for i, txt, in enumerate(rad2Detectors):
            ax. annotate(txt, (rad2Years[i], rad2Masses[i]), color='mediumorchid', fontsize=24, rotation=rotationvalue, va='top', ha='right')

   
    plt.plot([2017.5, 2017.5], [1, 5e13], color='gray', linestyle='--', linewidth=2)
            
    leg = ax.legend(loc='upper left', fontsize=24)
    leg.get_frame().set_alpha(0.0)
    ax.set_yscale('log')
    
    plt.ylim(1, 5e13)
    plt.ylabel('Active Volume Mass [kg]', fontsize=36)
    plt.yticks(fontsize=30)
    plt.xlim(1945, 2035)   
    plt.xlabel('', fontsize=30)
    plt.xticks(fontsize=24)
    

    plt.show()
    fig.savefig('./mass.png')
```
